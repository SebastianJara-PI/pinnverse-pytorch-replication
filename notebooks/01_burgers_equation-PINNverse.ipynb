# Burger's equation

Use the viscous form
\begin{align}
\dfrac{\partial u}{\partial t} + u \dfrac{\partial u}{\partial x} &= \nu \dfrac{\partial^2 u}{\partial x^2},\nonumber\\
u(0,x) &= -\sin(\pi x),\nonumber\\
u(t,x) &= 0 \,\,\,\,\,\,\hbox{at}\,\,\,\,\,\,x\in\{-1,1\},
\end{align}
where $u(t,x)$ represnet a velocity field. When $\nu > 0$, the solutions are smooth, balancing nonlinearity and diffusion. In the limit $\nu \rightarrow 0$, shock waves can form, making Burgers’ equation a prototypical model to study shock formation and related phenomena in fluid dynamics [1].

## Loss terms

```python
nu = nn.Parameter(torch.tensor(0.1, dtype=torch.float32))
```

```python
def burgers_residual():
    u = net(x_c, t_c)  
    u_t = torch.autograd.grad(u, t_c, torch.ones_like(u), create_graph=True)[0]
    u_x = torch.autograd.grad(u, x_c, torch.ones_like(u), create_graph=True)[0]
    u_xx = torch.autograd.grad(u_x, x_c, torch.ones_like(u_x), create_graph=True)[0]
    r = u_t + u*u_x - nu*u_xx
    return torch.mean(r**2)
```

```python
def initial_condition_loss():
    u_pred = net(x_i, t_i)
    return torch.mean((u_pred - u_i)**2)
```

```python
def boundary_condition_loss():
    u_b_l = net(x_b_l, t_b_l)
    u_b_r = net(x_b_r, t_b_r)
    return torch.mean(u_b_l**2) + torch.mean(u_b_r**2)
```

```python
def data_loss():
    u_data_1 = net(X_data, t_data_1)
    u_data_2 = net(X_data, t_data_2)
    r_data_1 = torch.mean((u_data_1 - U_data_1)**2)
    r_data_2 = torch.mean((u_data_2 - U_data_2)**2)
    return r_data_1 + r_data_2
```

## MDMM

```python
lam_f = nn.Parameter(torch.tensor(0.0, dtype=torch.float32))
lam_bc = nn.Parameter(torch.tensor(0.0, dtype=torch.float32))
lam_ic = nn.Parameter(torch.tensor(0.0, dtype=torch.float32))

rho_f, rho_bc, rho_ic = 1.0, 1.0, 1.0  # damping
```

```python
def pinnverse_total_loss():
    Ld = data_loss()
    
    Lf = burgers_residual()
    Lic = initial_condition_loss()
    Lbc = boundary_condition_loss()

    inf_Lf = -Lf
    inf_Lic = -Lic
    inf_Lbc = -Lbc

    J = Ld \
        + (lam_f * inf_Lf + 0.5* rho_f * inf_Lf**2) 
        # \
        # + (lam_ic * inf_Lic + 0.5* rho_ic * inf_Lic**2) \
        # + (lam_bc * inf_Lbc + 0.5* rho_bc * inf_Lbc**2)
    
    logs = {
        "Ld": Ld.detach(),
        "Lf": Lf.detach(),
        "Lbc": Lbc.detach(),
        "Lic": Lic.detach(),
        "inf_f": inf_Lf.detach(),
        "inf_bc": inf_Lbc.detach(),
        "inf_ic": inf_Lic.detach(),
        "lam_f": lam_f.detach(),
        "lam_bc": lam_bc.detach(),
        "lam_ic": lam_ic.detach(),
    }

    return J, logs
```

# Training

```python
X_test = torch.tensor(X0, dtype=torch.float32).reshape(-1, 1)
T_test = torch.tensor(T0, dtype=torch.float32).reshape(-1, 1)
U_true = torch.tensor(S.flatten(), dtype=torch.float32).reshape(-1, 1)

def relative_error():
    u_pred = net(X_test, T_test)
    return torch.norm(u_pred - U_true) / torch.norm(U_true)

#relative_error()
```

```python
dict = {
    "Ld": [],
    "Lf": [],
    "Lic": [],
    "Lbc": [],
    "inf_f": [],
    "inf_bc": [],
    "inf_ic": [],
    "lam_f": [],
    "lam_bc": [],
    "lam_ic": [],
    "J": [],
    "rel_error": [],
    "nu_value": []
}
```

```python
# tasa de aprendizaje distinta para pesos y para parametro nu
optimizer = torch.optim.Adam([
    {"params": net.parameters(), "lr": 1e-4},
    {"params": [nu], "lr": 1e-2},
    {"params": [lam_f, lam_bc, lam_ic], "lr": -1e-3}
])
#optimizer.param_groups[2]
```

```python
epoch = 10000
for ep in range(epoch):
    optimizer.zero_grad()
    J, logs = pinnverse_total_loss()
    J.backward()
    optimizer.step()

    dict["Ld"].append(logs["Ld"].item())
    dict["Lf"].append(logs["Lf"].item())
    dict["Lic"].append(logs["Lic"].item())
    dict["Lbc"].append(logs["Lbc"].item())
    dict["inf_f"].append(logs["inf_f"].item())
    dict["inf_bc"].append(logs["inf_bc"].item())
    dict["inf_ic"].append(logs["inf_ic"].item())
    dict["lam_f"].append(logs["lam_f"].item())
    dict["lam_bc"].append(logs["lam_bc"].item())
    dict["lam_ic"].append(logs["lam_ic"].item())
    dict["J"].append(J.item())
    dict["nu_value"].append(nu.item())

    if ep % 10 == 0:
        rel_err = relative_error().item()
        dict["rel_error"].append(rel_err)

    if ep % 1000 == 0:
        print(f"Epoch {ep}, nu: {nu.item():.6f} Relative Error: {rel_err:.6f}, Total Loss: {J.item():.6f}, Residual Loss: {logs['Lf'].item():.6f}, IC Loss: {logs['Lic'].item():.6f}, BC Loss: {logs['Lbc'].item():.6f}\n"
              f"inf_f: {logs['inf_f'].item():.6f}, inf_bc: {logs['inf_bc'].item():.6f}, inf_ic: {logs['inf_ic'].item():.6f}, lam_f: {logs['lam_f'].item():.6f}, lam_bc: {logs['lam_bc'].item():.6f}, lam_ic: {logs['lam_ic'].item():.6f}")
    
```
